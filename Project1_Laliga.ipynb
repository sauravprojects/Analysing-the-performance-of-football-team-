```python
#Part 4
#Converting GoalsFor and GoalsAgainst to integer type.
b["GoalsFor"] = b["GoalsFor"].astype(int)
b["GoalsAgainst"] = b["GoalsAgainst"].astype(int)
b.dtypes
```

```python
#Function to calculate Goals difference 
def Goals_diff_count(goalsfor,goalsagainst):
    return goalsfor- goalsagainst
x=Goals_diff_count(b["GoalsFor"],b["GoalsAgainst"])   
b["GoalDiff"]=x #Creating another column to display Goal Difference of each team
b
```

```python
b[b["GoalDiff"]==b["GoalDiff"].max()] #calculating maximum goal difference
```

```python
b[b["GoalDiff"]== b["GoalDiff"].min()] #minimum goal difference
```
